The dataset you will be working with includes intentional homicide and other crime rates in almost every country in the world over the past several years.
```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("SYB65_328_202209_Intentional homicides and other crimes.csv")
```

```python
df
```

```python
df = df.drop(df.index[0])
```

```python
df.columns.tolist()
```

```python
df.columns=['Region/Country/Area_id', 'Intentional homicides and other crimes', 'Year', 'Crime', 'Rate', 'Footnotes', 'Source']
```

```python
df
```

```python
df["Intentional homicides and other crimes"].unique()
```

```python
df["Intentional homicides and other crimes"].loc[]
```

```python
df = df.drop(df["Intentional homicides and other crimes"].index)
```

```python
countries=["Kenya", "Brazil", "Argentina", "Cabo Verde", "China"]
```

```python
df["Intentional homicides and other crimes"].isin(countries)
```

```python
df=df[df["Intentional homicides and other crimes"].isin(countries)]
```

```python
df["Crime"].unique()
```

```python
df
```

```python
df=df[df["Crime"].isin(["Intentional homicide rates per 100,000"])]
```

```python
df
```

```python
df=df[df["Year"].isin(["2015"])]
```

```python
df
```

```python
df["Rate"]=df["Rate"].astype(float)
```

```python
sns.barplot(data=df, x="Intentional homicides and other crimes", y="Rate")
```
