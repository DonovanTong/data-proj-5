### 1. 
Data visualization needs to be smart, or intellectually relevant to increase credibility. It also must be easy to understand. Lastly, it needs to be coherent. There shouldn't be any unnecessary patterns, colors, or text that is distracting and/or inconsistent.

### 2. 
- Order: a cultural norm by which we read information. E.g. left to right.

- Hierarchy: Certain aspects of an image stand out more. Light over dark, and red over grey are good examples.

- Clarity: Our brains expect simplicity. In order to help our audience understand, we need to keep our visualizations simple and clean.

- Relationships: Stories are best demonstrated through emphasizing relevant data interactions.

- Convention: A cultural standard that is determined as "right" or "proper".

- These are some key principles that will help to become a more credible, communicative, and trusted data analyst.

- Statistics is a tool that may be used to tell the truth only if one is precise and remains accurate to data. Otherwise, the whole truth is not told.

- One of my favorite misleading graphs I found was georgia's reported cases of covid 19. At first glance it seems as if the cases are decreasing rapidly in every county, but if you look closer, the x axis is set up in descending order of covid cases, not dates. We would see a much more accurate graph if the dates were in the correct order.

### 5. 

- Visualization clutter exists when certain conventions are not taken into consideration when creating a graph. Certain colors can inherently indicate certain states of the data such as good or bad.

- A graph includes these main elements: lines, labels, titles, and axis names.

- Less is more. Don't ignore convention. Accuracy and truthfulness are top priorities.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("salary_numeric.csv")
```

```python
df
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="industry_experience", y="salary")
plt.xlabel("Years of Experience")
plt.ylabel("Salary in USD")
plt.title("Manager Salaries by Years Experience", fontname="DejaVu Sans", fontweight="bold", fontdict={'fontsize':15, })
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="age", y="salary")
plt.xlabel("Age")
plt.ylabel("Salary in USD")
plt.title("Manager Salaries by Age", fontweight="bold", fontdict={'fontsize':15, })
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="age", y="industry_experience")
plt.xlabel("Age")
plt.ylabel("Experience")
plt.title("Age by Years Experience", fontname="DejaVu Sans", fontweight="bold", fontdict={'fontsize':15, })
```

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x="age", y="salary", line_kws={}, scatter_kws={"color":"#86c5dc"})
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Manager Salaries by Age", fontname="DejaVu Sans", fontweight="bold", fontdict={'fontsize':15, })
plt.yticks(np.arange(20000, 190000, step=10000))
```

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df, x="industry_experience", y="salary", line_kws={}, scatter_kws={"color":"#86c5dc"})
plt.xlabel("Experience in Years")
plt.ylabel("Salary")
plt.title("Manager Salaries by Experience", fontname="DejaVu Sans", fontweight="bold", fontdict={'fontsize':15, })
```
